```python
def load_json(fname):
    try:
        with open(fname, encoding="utf-8") as f:
            json_obj = json.load(f)
    except:
        with open(fname, encoding="utf-16") as f:
            json_obj = json.load(f)
    return json_obj

def write_json(data, fname):
    def _conv(o):
        if isinstance(o, (np.int64, np.int32)):
            return int(o)
        raise TypeError
    parent = os.path.dirname(fname)
    distutils.dir_util.mkpath(parent)
    try:
        with io.open(fname, "w", encoding="utf-8") as f:
            json_str = json.dumps(data, ensure_ascii=False, default=_conv)
            f.write(json_str)
    except:
        with io.open(fname, "w", encoding="utf-16") as f:
            json_str = json.dumps(data, ensure_ascii=False, default=_conv)
            f.write(json_str)

def make_result(data_path, save_path):
    csvfiles, result = [], []
    for file in glob.glob(data_path):
        csvfiles.append(file)
    for i in csvfiles:
        print(i)
        result.extend(load_json(i))
    if len(result)==23015 or len(result) == 10740:
        write_json(result, save_path)
        print(len(result), 'file saved')
    else:
        print('len of result is not 23015 or 10740 : {}'.format(len(result)))
```
